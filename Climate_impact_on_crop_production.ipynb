```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.dropna()
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['Crop', 'Season', 'State'])
```

```python
X = df.drop(['Production', 'Yield'], axis=1)
y_production = df['Production']
y_yield = df['Yield']
```

```python
df.head()
```

```python
#
X_train, X_test, y_production_train, y_production_test = train_test_split(X, y_production, test_size=0.2, random_state=42)
_, _, y_yield_train, y_yield_test = train_test_split(X, y_yield, test_size=0.2, random_state=42)
```

```python
model_production = LinearRegression()
model_yield = LinearRegression()
```

```python
model_production.fit(X_train, y_production_train)
model_yield.fit(X_train, y_yield_train)
```

```python
y_production_pred = model_production.predict(X_test)
y_yield_pred = model_yield.predict(X_test)
```

```python
mse_production = mean_squared_error(y_production_test, y_production_pred)
r2_production = r2_score(y_production_test, y_production_pred)
mse_yield = mean_squared_error(y_yield_test, y_yield_pred)
r2_yield = r2_score(y_yield_test, y_yield_pred)
```

```python
print(f'Production Model - Mean Squared Error: {mse_production}, R2 Score: {r2_production}')
print(f'Yield Model - Mean Squared Error: {mse_yield}, R2 Score: {r2_yield}')
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 2, 1)
sns.scatterplot(x=y_production_test, y=y_production_pred)
plt.xlabel('Actual Production')
plt.ylabel('Predicted Production')
plt.title('Actual vs Predicted Production')
```

```python
plt.subplot(1, 2, 2)
sns.scatterplot(x=y_yield_test, y=y_yield_pred)
plt.xlabel('Actual Yield')
plt.ylabel('Predicted Yield')
plt.title('Actual vs Predicted Yield')
```

```python
plt.subplot(1, 2, 2)
sns.scatterplot(x=y_yield_test, y=y_yield_pred)
plt.xlabel('Actual Yield')
plt.ylabel('Predicted Yield')
plt.title('Actual vs Predicted Yield')
plt.tight_layout()
plt.show()
```
